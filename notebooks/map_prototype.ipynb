```python
import pandas as pd
import folium

clean_df = pd.read_csv(
    "../data/clean_parking_data.csv"
)
```

```python
stations = sorted(
    clean_df["police_station"]
    .dropna()
    .unique()
)

len(stations)
```

```python
station = clean_df["police_station"].unique()[0]
hour = clean_df["hour"].unique()[0]


station_df = clean_df[
    clean_df["police_station"] == station
]

station_df.shape
```

```python


hour_df = station_df[
    station_df["hour"] == hour
]

hour_df.shape
```

```python
sample_df = hour_df.head(500)
```

```python
m = folium.Map(
    location=[12.9716,77.5946],
    zoom_start=11
)
```

```python
for _, row in sample_df.iterrows():

    popup_text = f"""
                <b>Vehicle:</b> {row['vehicle_type']}<br>
                <b>Violation:</b> {row['primary_violation']}<br>
                <b>Station:</b> {row['police_station']}<br>
                <b>Hour:</b> {row['hour']}<br>
                <b>Junction:</b> {row['junction_name']}
                """

    folium.CircleMarker(
        location=[
            row["latitude"],
            row["longitude"]
        ],
        radius=3,
        color="blue",
        fill=True,
        fill_color="blue",
        fill_opacity=0.7,
        popup=popup_text
    ).add_to(m)
```
