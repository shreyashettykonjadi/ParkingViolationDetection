```python
import pandas as pd

df = pd.read_csv("../data/clean_parking_data.csv")

SHIFT_BINS = [0, 6, 9, 14, 20, 24]
SHIFT_LABELS = ["night", "morning_peak", "midday", "evening_peak", "late_night"]

df["shift"] = pd.cut(df["hour"], bins=SHIFT_BINS, labels=SHIFT_LABELS, right=False)
df["day_type"] = df["weekday"].apply(lambda x: "weekend" if x >= 5 else "weekday")

df.groupby(["day_type", "shift"], observed=True).size().reset_index(name="record_count")
```
