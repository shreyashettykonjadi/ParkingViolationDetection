```python
import pandas as pd
import numpy as np
import ast
```

```python
df = pd.read_csv("../data/parking_data.csv")
```

```python
clean_df = df.copy()
```

```python
clean_df = clean_df.drop(
    columns=[
        "description",
        "closed_datetime",
        "action_taken_timestamp"
    ]
)
```

```python
clean_df["created_datetime"] = pd.to_datetime(
    clean_df["created_datetime"],
    format="mixed",
    utc=True
)
```

```python
clean_df["hour"] = (
    clean_df["created_datetime"]
    .dt.hour
)
```

```python
clean_df["weekday"] = (
    clean_df["created_datetime"]
    .dt.dayofweek
)
```

```python
clean_df["month"] = (
    clean_df["created_datetime"]
    .dt.month
)
```

```python
clean_df["date"] = (
    clean_df["created_datetime"]
    .dt.date
)
```

```python
clean_df["violation_type"] = (
    clean_df["violation_type"]
    .apply(ast.literal_eval)
)
```

```python
def is_parking_violation(v):
    return any(
        "PARKING" in item
        for item in v
    )
```

```python
clean_df["is_parking_violation"] = (
    clean_df["violation_type"]
    .apply(is_parking_violation)
)
```
